```python
import datetime
 
# Function to convert string to datetime
def convert(date_time):
    format = '%d-%b-%Y'  # The format
    datetime_str = datetime.datetime.strptime(date_time, format)
 
    return datetime_str
```

```python
import datetime
import calendar
 
def findDay(date):
    born = datetime.datetime.strptime(date, '%Y-%m-%d').weekday()
    return born
```
